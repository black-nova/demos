```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop(['Prediction','Email No.'],axis=1))
```

```python
scaled_features=scaler.transform(df.drop(['Prediction','Email No.'],axis=1))
temp_1=df['Email No.']
df.drop('Email No.',axis=1,inplace=True)
df['Email No.']=temp_1
df_id=pd.DataFrame(scaled_features,columns=df.columns[:-2])
df_id
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_id
y=df['Prediction']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.confusion_matrix(y_test,pred))
print(metrics.classification_report(y_test,pred))
```

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,40),error_rate,marker='o',markerfacecolor='red',ls='--',lw='3',markersize=10)
```
